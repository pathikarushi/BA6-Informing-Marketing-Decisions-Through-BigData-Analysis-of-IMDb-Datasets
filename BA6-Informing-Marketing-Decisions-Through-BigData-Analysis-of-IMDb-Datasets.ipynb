### 1.2 Description of Data and its Source <a class="anchor" id="Data-description-source"></a> 

IMDb, short for the Internet Movie Database, is an online database that provides comprehensive information on films, television programs, and video games, along with cast and crew details, ratings, reviews, and more. Launched in 1990 by Col Needham, IMDb has grown into one of the most trusted and reliable sources of entertainment data, serving millions of users worldwide.

The IMDb dataset is an extensive collection of data, which is meticulously curated and regularly updated by IMDb staff, industry professionals, and a passionate community of users. The dataset is derived from various sources such as production companies, distributors, filmmakers, and individual contributors, ensuring a diverse and accurate representation of the entertainment industry.

IMDb datasets encompass multiple categories, including:

**Title information:** Details about films, TV shows, and video games, including title, release year, runtime, and genre.

**Cast and crew information:** Comprehensive data on actors, directors, writers, and other crew members involved in a production.

**Ratings and reviews:** User-generated ratings and reviews, providing insight into the reception and popularity of a title.

**Trivia and goofs:** Interesting facts, behind-the-scenes information, and any errors or inconsistencies found within the content.

**Technical specifications:** Information on aspects like filming locations, production budgets, box office revenues, and technical details like aspect ratio and sound formats.

IMDb data offers a wealth of information that can be invaluable to researchers, marketers, content creators, and entertainment enthusiasts alike. By tapping into this rich dataset, users can gain a deeper understanding of the film and television industry, identify trends and patterns, and make well-informed decisions based on data-driven insights.

### 1.3 IMDb Dataset Manual <a class="anchor" id="Dataset-Manual"></a>

**Data Location**

The dataset files can be accessed and downloaded from https://datasets.imdbws.com/. The data is refreshed daily.

**IMDb Dataset Details**

Each dataset is contained in a gzipped, tab-separated-values (TSV) formatted file in the UTF-8 character set. The first line in each file contains headers that describe what is in each column. The datasets are as follows:

**title.akas.tsv.gz** - Contains the following information for titles:

* titleId (string) - a tconst, an alphanumeric unique identifier of the title
* ordering (integer) – a number to uniquely identify rows for a given titleId
* title (string) – the localized title
* region (string) - the region for this version of the title
* language (string) - the language of the title
* types (array) - Enumerated set of attributes for this alternative title. One or more of the following: "alternative", "dvd", "festival", "tv", "video", "working", "original", "imdbDisplay". New values may be added in the future without warning
* attributes (array) - Additional terms to describe this alternative title, not enumerated
* isOriginalTitle (boolean) – 0: not original title; 1: original title

**title.basics.tsv.gz** - Contains the following information for titles:

* tconst (string) - alphanumeric unique identifier of the title
* titleType (string) – the type/format of the title (e.g. movie, short, tvseries, tvepisode, video, etc)
* primaryTitle (string) – the more popular title / the title used by the filmmakers on promotional materials at the point of release
* originalTitle (string) - original title, in the original language
* isAdult (boolean) - 0: non-adult title; 1: adult title
* startYear (YYYY) – represents the release year of a title. In the case of TV Series, it is the series start year
* endYear (YYYY) – TV Series end year. ‘\N’ for all other title types
* runtimeMinutes – primary runtime of the title, in minutes
* genres (string array) – includes up to three genres associated with the title

**title.crew.tsv.gz** – Contains the director and writer information for all the titles in IMDb. Fields include:

* tconst (string) - alphanumeric unique identifier of the title
* directors (array of nconsts) - director(s) of the given title
* writers (array of nconsts) – writer(s) of the given title

**title.episode.tsv.gz** – Contains the tv episode information. Fields include:

* tconst (string) - alphanumeric identifier of episode
* parentTconst (string) - alphanumeric identifier of the parent TV Series
* seasonNumber (integer) – season number the episode belongs to
* episodeNumber (integer) – episode number of the tconst in the TV series

**title.principals.tsv.gz** – Contains the principal cast/crew for titles

* tconst (string) - alphanumeric unique identifier of the title
* ordering (integer) – a number to uniquely identify rows for a given titleId
* nconst (string) - alphanumeric unique identifier of the name/person
* category (string) - the category of job that person was in
* job (string) - the specific job title if applicable, else '\N'
* characters (string) - the name of the character played if applicable, else '\N'

**title.ratings.tsv.gz** – Contains the IMDb rating and votes information for titles

* tconst (string) - alphanumeric unique identifier of the title
* averageRating – weighted average of all the individual user ratings
* numVotes - number of votes the title has received

**name.basics.tsv.gz** – Contains the following information for names:

* nconst (string) - alphanumeric unique identifier of the name/person
* primaryName (string)– name by which the person is most often credited
* birthYear – in YYYY format
* deathYear – in YYYY format if applicable, else '\N'
* primaryProfession (array of strings)– the top-3 professions of the person
* knownForTitles (array of tconsts) – titles the person is known for

#### 4.2.1 What is the distribution of the genres from year 1874 to 2023? <a class="anchor" id="dist-genres"></a>  

```python
from pyspark.sql.functions import split, explode, concat_ws

# Convert the 'genres' column from array<string> to string type
title_basics = title_basics.withColumn('genres_str', concat_ws(',', 'genres'))

# Split the 'genres_str' column by comma and explode the resulting array
split_genres = title_basics.select('startYear', 'runtimeMinutes', explode(split('genres_str', ',')).alias('genre'))

# Count the number of occurrences of each genre and plot the distribution
genre_counts = split_genres.groupBy('genre').count().orderBy('count', ascending=False)
df = genre_counts.toPandas()
df = df.set_index('genre')
df['count'].plot.bar(figsize=(10, 6))
plt.title('Genres Distribution')
plt.xlabel('Genres')
plt.ylabel('Counts');
```

#### 4.2.2 Which genre is the most popular each year? <a class="anchor" id="popular-genres"></a>  

```python
from pyspark.sql.functions import explode, row_number, desc
from pyspark.sql.window import Window
import matplotlib.pyplot as plt

# Explode the 'genres' column to split it into multiple rows
exploded_genres = title_basics.select('startYear', 'runtimeMinutes', explode('genres').alias('genre'))

# Group by 'startYear', 'runtimeMinutes', and 'genre' to count the occurrences of each genre in each year
genre_counts_by_year = exploded_genres.groupBy('startYear', 'genre').count()

# Define a window specification to rank the genres by count within each year
window_spec = Window.partitionBy('startYear').orderBy(desc('count'))

# Filter to the top-ranked genre in each year from 2000 to 2022
top_genre_by_year = (genre_counts_by_year
                     .withColumn('rank', row_number().over(window_spec))
                     .filter('rank = 1')
                     .filter((genre_counts_by_year['startYear'] >= 2000) & (genre_counts_by_year['startYear'] <= 2022))
                     .orderBy('startYear'))

# Convert to a Pandas dataframe and plot the results
top_genre_by_year_df = top_genre_by_year.toPandas()
plt.figure(figsize=(10, 6))
plt.bar(top_genre_by_year_df['startYear'], top_genre_by_year_df['count'], color='skyblue')
plt.title('Top Genre by Year (2000-2022)')
plt.xlabel('Year')
plt.ylabel('Count')
for i, row in top_genre_by_year_df.iterrows():
    plt.annotate(row['genre'], (row['startYear'], row['count']), ha='center', va='bottom')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

**Answer:**

The graphs reveal that between 2000 and 2022, drama consistently ranks as the most popular genre released each year, with the exceptions of 2016 and 2020 when comedy takes the top spot. We chose to focus on the leading genre annually from 2000 to 2022 because people's preferences have evolved over time, rendering data from before 2000 less pertinent. Furthermore, the dataset is incomplete, which led us to analyze the years 2000 through 2022 specifically.

#### 6.6.1 Hyper Parameter Tuning <a class="anchor" id="6-6-1"></a>

```python
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator
from pyspark.ml.evaluation import RegressionEvaluator

# Create a vector assembler to assemble the features
assembler = VectorAssembler(inputCols=['startYear', 'runtimeMinutes', 'popular_actor_flag', 
                                       'popular_director_flag', 'popular_writer_flag'], outputCol='features')

# Assemble the features and scale them using StandardScaler
scaled_data = StandardScaler(inputCol='features', outputCol='scaledFeatures').fit(assembler.transform(dataset)).transform(assembler.transform(dataset))

# Split the dataset into training and test sets
train, test = scaled_data.randomSplit([0.8, 0.2], seed=13)

# Create a linear regression object
lr = LinearRegression(featuresCol='scaledFeatures', labelCol='popularityIndex')

# Create a parameter grid builder
param_grid = ParamGridBuilder() \
    .addGrid(lr.regParam, [0.0, 0.01, 0.1]) \
    .addGrid(lr.elasticNetParam, [0.0, 0.5, 1.0]) \
    .build()

# Create a cross-validator object
cv = CrossValidator(estimator=lr, estimatorParamMaps=param_grid, evaluator=RegressionEvaluator(labelCol='popularityIndex', predictionCol='prediction', metricName='rmse'), numFolds=5, seed=123)

# Train the model and select the best set of parameters
cv_model = cv.fit(train)

# Make predictions on the test set and evaluate the model
predictions = cv_model.transform(test)
evaluator = RegressionEvaluator(labelCol='popularityIndex', predictionCol='prediction', metricName='rmse')
rmse = evaluator.evaluate(predictions)

print(f'LinReg CV RMSE on test set: {rmse:.3f}')
```
